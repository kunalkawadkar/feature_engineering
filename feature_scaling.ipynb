# standardization

```python
import seaborn as sns
```

```python
df =sns.load_dataset('tips')
```

```python
df.head()
```

```python
total_bills = list(df['total_bill'])
```

```python
total_bills
```

```python
sns.histplot(total_bills)
```

```python
import numpy as np
mean = np.mean(total_bills)
```

```python
std = np.std(total_bills)
```

```python
mean,std
```

```python
normalized_data =[]
for i in total_bills:
  z_score = (i-mean)/std
  normalized_data.append(z_score)
```

```python
z_score
```

```python
normalized_data
```

```python
sns.histplot(normalized_data)
```
